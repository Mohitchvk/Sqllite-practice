```python
from sqlite3 import Error


def create_connection(db_file):
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
    finally:
        if conn:
            conn.close()


create_connection("pythonsqlite.db")
```

```python
conn = sqlite3.connect("pythonsqlite.db")
cursor = conn.cursor()
```
